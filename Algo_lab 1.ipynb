```python
def MinDestance(a):
    dmin = abs(a[0] - a[1])
    for i in range(0,len(a)):
        for j in range(2,len(a)):
            if (i != j and abs(a[i] - a[j]) < dmin):
                dmin = abs(a[i] - a[j])
    return dmin
```

```python
a  = [1,2,3,4,5]
a
```

```python
MinDestance(a)
```

```python
%ti
```

```python
%timeit MinDestance(a)
```

```python
def MinDestance2(a):
    dmin = abs(a[0] - a[1])
    for i in range(0,len(a)):
        for j in range((i + 1),len(a)):
            if (i != j and abs(a[i] - a[j]) < dmin):
                dmin = abs(a[i] - a[j])
    return dmin
```

```python
%timeit MinDestance2(a)
```

```python
MinDestance2(a)
```

```python
def MinDestance3(a):
    dmin = abs(a[0] - a[1])
    n = len(a)
    for i in range(0,n-1):
        for j in range((i + 1),n):
            if (abs(a[i] - a[j]) < dmin):
                dmin = abs(a[i] - a[j])
    return dmin
```

```python
%timeit MinDestance3(a)
```

```python
MinDestance3(a)
```

```python
b = [5,3,1,8,9]
```

```python
def MinDestance4(b):
    b.sort()
    dmin = abs(b[0] - b[1])
    for i in range(3,len(b)):
        if(dmin > abs(b[i] - b[i-1])):
            dmin = abs(b[i] - b[i-1])
    return dmin
```

```python
%timeit MinDestance4(b)
```

```python
MinDestance4(b)
```
